```python
import numpy as np
import sys
from Port_modelClass import CreditPort
```

# Simple Input Samples

```python
PDs = [0.5,0.1,0.2]
RR_dists = [[0.2, 0.3, 0.5], [0.9, 0.05,0.05],[0.4, 0.3, 0.3]]
RR_intervals = [0, 0.2, 0.6, 1]
exposures = [1 for i in range(0, len(PDs))]
correlation_matrix = np.ones((len(PDs), len(PDs)))*0.2
np.fill_diagonal(correlation_matrix, 1)
print('\nPDs:\n',PDs,'\nRR_dists:\n',RR_dists,'\nRR_intervals:\n',RR_intervals,'\ncorrelation_matrix:\n',correlation_matrix)
```

```python
myPort = CreditPort(PDs = PDs, RR_intervals = RR_intervals, RR_dists = RR_dists, exposures = exposures , correlation_matrix=correlation_matrix)
myPort.genKPI()
```

```python
myPort.simulate_portfolio()
myPort.genKPI()
```

```python
myPort.simulate_portfolio(10000, 't', t_v=4)
myPort.genKPI()
```
