```python
import ML_models as Models
import transformationFunction as Tff
import pandas as pd
import numpy as np
from random import gauss
import matplotlib.pyplot as plt
import math
```

# Sample 1
The most simple sample to use linear model without target transformation.
## Generate observation data 
We generate a simple data with a trend.

```python
data_length=100
Observations = pd.Series([gauss(0, 5) + i for i in range(0, data_length)])
Observations.name = 'Observations'
```

```python
plt.plot(Observations)
```

## Initialization
We use the identity function as the transformation Function, which means it does not change anything.

```python
mModel = Models.linearModel(Observations,Tff.identity())
```

## xAndPart
X is the Explantory Data. We can partition it and fill the missing data with some constants

```python
x = pd.DataFrame({'feature1': [20*i+100 for i in range(0, data_length)]})
x.loc[[20,66,34],:] = np.nan
mModel.xAndPart(x, partitionRatio=0.8, fillingValue=1000)
plt.plot(x)
```

## Fit

```python
mModel.fit()
```

## Error Revalution

```python
print(f"The training MAE of the model and naive model are {mModel.trainError('MAE')}")
```

```python
print(f"The testing MAE of the model and naive model are {mModel.testError('MAE')}")
```

### Predict

```python
mModel.predict(x[3:7])
```

# Sample 2
This sample is for classfication problem which use logistic models without transformation.
## Generate observation data and initilization the model
We generate a classfication data with probability trend.

```python
data_length=100
Observations = (pd.Series([gauss(0, 30) + i for i in range(0, data_length)])>50).astype(int)
Observations.name = 'Observations'
print(Observations[45:55])
plt.plot(Observations,'.')
```

```python
mModel = Models.logisticRegression(Observations, Tff.identity())
```

## xAndPart, fit and revalution
X is the Explantory Data. We can try differnt x as the features.   
Here, we try sin and cos as features firstly, which are definitely not good for our simulated data. Then we try the linear increasing feature, which should output good predictions because our data has a trend.

```python
x1 = pd.DataFrame({'feature1': [math.sin(i) for i in range(0, data_length)],
                   'feature2': [math.cos(i) for i in range(0, data_length)]})
mModel.xAndPart(x1, partitionRatio=0.6)
print(mModel.x_tr[0:5])
print(mModel.y_tr[0:5])
```

```python
mModel.fit()
print(f"The training AUC of the model and naive model are {mModel.trainError('AUC')[0]}")
print(f"The testing AUC of the model and naive model are {mModel.testError('AUC')[0]}")
plt.plot(mModel.testError('AUC')[1],mModel.testError('AUC')[2])
plt.plot([0,1],[0,1],'--')
```

These features obviously do not perform well. Hence, we may try to find other features for machine learning.

```python
x2 = pd.DataFrame({'feature1': [i for i in range(0, data_length)]})
mModel.xAndPart(x2, partitionRatio=0.6)
```

```python
mModel.fit()
print(f"The training AUC of the model and naive model are {mModel.trainError('AUC')[0]}")
print(f"The testing AUC of the model and naive model are {mModel.testError('AUC')[0]}")
plt.plot(mModel.testError('AUC')[1],mModel.testError('AUC')[2])
plt.plot([0,1],[0,1],'--')
```

These features perfrom much better! This process shows the usage of the model on feature selection.

### Predict

```python
print(mModel.predict(x2[int(data_length/2)-10:int(data_length/2)+10]))
print(mModel.predictProb(x2[2:8]))
```

# Sample 3
This sample will show a regression with transformation functions.
## Generate observation data 
We simulate a "U-shape" data which has a probability data to predict. 

```python
data_length=200
Observations = pd.Series(1/(1+ np.exp(0.1*np.array([gauss(0, 30) + i -data_length / 2 for i in range(0, data_length)]))))
Observations.name = 'Observations'
plt.plot(Observations,'.')
plt.show()
plt.hist(Observations)
```

## Initialization
We initialize two linear models whose only difference is the tranforming function.

```python
mModel1 = Models.linearModel(Observations,Tff.identity())
mModel2 = Models.linearModel(Observations,Tff.sigmoid())
```

## xAndPart, fit and revalution

```python
x = pd.DataFrame({'feature1': [i for i in range(0, data_length)]})
mModel1.xAndPart(x)
mModel2.xAndPart(x)
mModel1.fit()
mModel2.fit()
print(f"The testing MAE of the model and naive model are {mModel1.testError('MAE')}")
print(f"The testing MAE of the model and naive model are {mModel2.testError('MAE')}")
```

From the results, we can see the impact of the transformation function for some specific data.

```python
### Predict
```

```python
mModel2.predict(x[3:7])
```

# Sample 4
This example is for the presentation of the usage of "segments" function.  
This problem is a regression problem, but we use the "segments" function as a classfication problem. Hence, the ways to change the 'class' prediction to a 'value' prediction can be different. The 'mean' way is to calculate the product of the probability vector and class number middle vector. The max way is find the class with the highest probability and transform back to the middle value of the class.  

```python
data_length=2000
Observations = pd.Series(1/(1+ np.exp(0.005*np.array([gauss(0, 30) + i -data_length / 2 for i in range(0, data_length)]))))
Observations.name = 'Observations'
plt.plot(Observations,'.')
plt.show()
plt.hist(Observations)
```

```python
mModel = Models.CalibratedRandomForestC(Observations,Tff.segment(6))
```

```python
x = pd.DataFrame({'feature1': [i for i in range(0, data_length)]})
mModel.xAndPart(x)
mModel.fit()
print(f"The testing MAE of the model and naive model are {mModel.testError('MAE')}")
print(f"The testing MAE of the model and naive model are {mModel.testError('maxMAE')}")
```

### Predict

```python
print(mModel.predict(x[3:7], class2valueFun='mean'))
print(mModel.predict(x[3:7], class2valueFun='max'))
```

```python
print(mModel.predictProb(x[int(data_length/2)-3:int(data_length/2)+3]))
```
